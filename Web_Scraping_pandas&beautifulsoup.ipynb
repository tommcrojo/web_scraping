# Find and Find_All

```python
from bs4 import BeautifulSoup 
import requests
```

```python
url="https://www.scrapethissite.com/pages/forms/"
```

```python
page= requests.get(url)
soup=BeautifulSoup(page.text,"html")
```

```python
# find : finds the first result that has the query we look for
# find_all : finds all results that match the search
soup.find_all("div")
```

```python
soup.find_all("p",class_="lead")
```

### How to extract text from a webpage using Python

```python
# find_all doesn't have text attribute, you must use find
soup.find_all("p",class_="lead").text
```

```python
soup.find("p",class_="lead").text.strip()
```

### Extract Tables from website in Python

```python
soup.find_all("th")
```

```python
soup.find("th").text.strip()
```

# Real Website Extraction + Pandas

```python
from bs4 import BeautifulSoup 
import requests
```

```python
url="https://en.wikipedia.org/wiki/List_of_largest_companies_in_the_United_States_by_revenue"
page=requests.get(url)
soup=BeautifulSoup(page.text,"html")
```

```python
soup.find_all("table")[1]
```

```python
table=soup.find_all("table")[1]
```

```python
world_titles=table.find_all("th")
```

```python
world_table_titles=[title.text.strip() for title in world_titles]
```

```python
print(world_table_titles)
```

- th = table header
- tr = table row
- td = table data

```python
import pandas as pd
```

```python
df= pd.DataFrame(columns=world_table_titles)
df
```

```python
column_data=table.find_all("tr")
```

```python
for row in column_data[1:]:
    row_data = row.find_all("td")
    individual_row_data=[data.text.strip() for data in row_data]
    length=len(df)
    df.loc[length] = individual_row_data
```
